# Imports

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 2D plotting using Matplotlib, MATLAB Style
`plt.plot` documentation here: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html

```python
x = np.linspace(0.0, 2.0, 20)
plt.plot(x, np.sqrt(x), 'ro')
plt.show()
```

## Blue line

```python
plt.plot(x, np.sqrt(x), 'b-')
plt.show()
```

## Three lines on the same figure

```python
plt.plot(x, x, 'g--', label='green') # green dashed line
plt.plot(x, np.sqrt(x), 'ro', label='red') # red circles
plt.plot(x, np.sqrt(x), 'b-', label='blue') # blue line
plt.legend()
plt.show()
```

# Multiple plots in the same figure
see https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html

```python
plt.subplot(1,2,1)
plt.plot(x, x, 'g--')
plt.subplot(1,2,2)
plt.plot(x, np.sqrt(x), 'b-')
plt.show()
```

## Save a Figure to a File

```python
plt.plot(x, np.sqrt(x), 'b-')

plt.savefig('myfig.png') # can also save to pdf by changing the file extension
```

# Matplotlib object oriented style

from https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html

```python
from scipy.stats import norm

x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
r = norm.rvs(size=1000)
rv = norm()

fig, ax = plt.subplots(1, 1)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='pdf')
ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
```

# Plotting from a DataFrame
See: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.plot.html

```python
names = ['Ceres', 'Pluto', 'Haumea', 'Makemake', 'Eris']
distance = [2.77, 39.26, 43.13, 45.79, 68.01] #in AU
dwarf_planets = list(zip(names, distance))
dp = pd.DataFrame(data=dwarf_planets, columns=['name', 'distance'])
dp
```

Plot the "distance" column

```python
dp['distance'].plot()
plt.show()
```

# Style Sheets
See: https://matplotlib.org/gallery/style_sheets/style_sheets_reference.html

```python
plt.style.available[:4]
```

```python
with plt.style.context('fivethirtyeight'):
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
    plt.show()
```

```python
with plt.style.context('seaborn'):
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
    plt.show()
```

# XKCD Easter Egg

```python
# https://matplotlib.org/xkcd/examples/showcase/xkcd.html

def xkcd_fun():
    plt.xkcd()
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    plt.xticks([])
    plt.yticks([])
    ax.set_ylim([-30, 10])
    data = np.ones(100)
    data[70:] -= np.arange(30)
    plt.plot(data)

    plt.annotate('THE DAY I REALIZED\nI COULD COOK BACON\nWHENEVER I WANTED',
        xy=(70, 1), arrowprops=dict(arrowstyle='->'), xytext=(15, -10))

    plt.xlabel('time')
    plt.ylabel('my overall health')
    plt.show()
```

```python
xkcd_fun()
```
