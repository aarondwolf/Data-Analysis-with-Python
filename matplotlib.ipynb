# Matplotlib

Matplotlib is the standard plotting package for Python. With enough man-handling, it can makes plots that look pretty much anyway you want.

It has a lot of really great examples at http://matplotlib.org/gallery.html

## 2D plotting using Matplotlib

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0.0, 2.0, 20)
```

##  Red circles

```python
plt.plot(x, np.sqrt(x), 'ro')
plt.show()
```

## Blue line

```python
plt.plot(x, np.sqrt(x), 'b-')
plt.show()
```

## Three lines on the same figure

```python
plt.plot(x, x, 'g--', label='green') # green dashed line
plt.plot(x, np.sqrt(x), 'ro', label='red') # red circles
plt.plot(x, np.sqrt(x), 'b-', label='blue') # blue line
plt.show()
```

# Multiple plots in the same figure

```python
plt.subplot(1,2,1)
plt.plot(x, x, 'g--')
plt.subplot(1,2,2)
plt.plot(x, np.sqrt(x), 'b-')
plt.show()
```

## Save a Figure to a File

```python
plt.plot(x, np.sqrt(x), 'b-')

plt.savefig('myfig.png') # can also save directly to pdf by just changing the file extension
```

# Plotting from a DataFrame

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
names = ['Ceres', 'Pluto', 'Haumea', 'Makemake', 'Eris']
distance = [2.77, 39.26, 43.13, 45.79, 68.01] #in AU
dwarf_planets = list(zip(names, distance))
dwarf_planets
```

Convert the list of tuples into a Data Frame

```python
dp = pd.DataFrame(data=dwarf_planets, columns=['name', 'distance'])
dp
```

Plot the "distance" column

```python
dp.plot()
```
